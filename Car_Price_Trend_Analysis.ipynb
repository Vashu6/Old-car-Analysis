```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel("/content/drive/MyDrive/carPriceTrendAnalysis.xlsx")
```

```python
df.head()
```

```python
df.dtypes
```

```python
df=df.drop(columns='Unnamed: 0',axis=1)
```

```python
df['kms_driven']=df['kms_driven'].str.replace("kms","")
```

```python
df['ownership']=df['ownership'].str.replace("st","").str.replace("Owner"," ").str.replace("nd","").str.replace("rd","").str.replace("th","")
```

```python
df['engine']=df['engine'].str.replace("cc","")
```

```python
df['Seats']=df['Seats'].str.replace("Seats","")
```

```python
df
```

```python
df=df.dropna()
```

```python
for col in ['kms_driven', 'ownership', 'engine', 'Seats']:
    try:
        df[col] = df[col].astype('int')
    except:
        df[col] = df[col]
```

```python
m=[]
for i in df['kms_driven']:
  d=i.replace(',','')
  a=d.strip()
  m.append(int(d))
print(m)
```

```python
df['kms_driven']=m
```

```python
df.info()
```

```python
main=[]
for i in df['car_prices_in_rupee']:
  if 'Crore' in str(i):
    c=i.replace('Crore','')
    main.append(int(float(c)*100)*100000)
  elif 'Lakh' in str(i):
    l=i.replace('Lakh','')
    main.append(int(float(l)*100000))
  else:
    main.append(i)
print(main)
```

```python
df['car_prices_in_rupee']=main
```

```python
df['car_prices_in_rupee'] = df['car_prices_in_rupee'].astype('int')
```
